```python
from nltk.tokenize import word_tokenize
from tensorflow.keras.preprocessing.text import one_hot
voc_size = 10000
input = []
for x in data['Sentences']:
    val = word_tokenize(x) # convert each sentence to separate list of word , separated by comma.
    one_rpr = one_hot(str(val),n=voc_size,lower=False,split=None) # encode each list of words to numerical value
    input.append(one_rpr) # appending and making a separate list of numerical value
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length = 15 # length of words model will look for training as input data.
embedded_docs = pad_sequences(input,padding='post',maxlen=sent_length) 
#padding is 'post' - so after 15 words, it will add 0 so data will be of same length. we can also set it for 'pre'.
```

```python
model =Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , precision_score , recall_score , f1_score , roc_auc_score
def Evaluation_metrics(actual,pred):
    accuracy_sco = accuracy_score(actual,pred)
    confusion_mat = confusion_matrix(actual,pred)
    precision_sco = precision_score(actual,pred,average='macro')
    recall_sco = recall_score(actual,pred,average='macro')
    f1_sco = f1_score(actual,pred,average='macro')
    return accuracy_sco,confusion_mat,precision_sco,recall_sco,f1_sco
```

```python
Evaluation_metrics(y_test,predicted_val)
```

```python
# Past - 1 , Present - 2  , future - 0
# created fn where we can give sentenec and it will do all process to change into numerical sentence and it will change its shape.
# and predict and return which category it belongs
from nltk.tokenize import word_tokenize
from tensorflow.keras.preprocessing.text import one_hot
def try_other_sentences(sentence):
    voc_size_val = 10000
    input_val = [sentence] 
    out_put = []
    for x in input_val:
        val = word_tokenize(x)
        one_rpr = one_hot(str(val),n=voc_size_val,lower=False,split=None)
        out_put.append(one_rpr)
    sent_length = 15
    embedded_sentence = pad_sequences(out_put,padding='post',maxlen=sent_length) #input- sentence
    value_to_pass = model.predict(embedded_sentence) 
    rows = value_to_pass.shape[0]
    cols = value_to_pass.shape[1] * value_to_pass.shape[2]
    value_to_pass = value_to_pass.reshape(rows,cols) #coverting test sentence to same shape of 'x' . as we trained on.
    return value_to_pass
def predict_output(value):
    pred = knn.predict(value)
    if pred == [1]:
        print('Past Tense',pred)
    elif pred == [2]:
        print('Present Tense',pred)
    elif pred == [0]:
        print('Future Tense',pred)
```
