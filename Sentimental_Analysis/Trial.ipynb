```python
from tensorflow.keras.layers import SimpleRNN, Dense

def create_simple_rnn_model(input_shape, num_units, num_classes):
    rnn_model = Sequential()
    rnn_model.add(SimpleRNN(units=num_units, input_shape=input_shape))
    rnn_model.add(Dense(units=num_classes, activation='softmax'))
    rnn_model.add(Dense(units=1, activation='softmax')) # or 'sigmoid' for binary
    rnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) # or 'binary_crossentropy'
    return rnn_model
```

```python
timesteps = 20
features = 15
num_classes = 3
num_units = 32
batch_size = 2769
input_shape = (timesteps, features)

model = create_simple_rnn_model(input_shape, num_units, num_classes)
```
